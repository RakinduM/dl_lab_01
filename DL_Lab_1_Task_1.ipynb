```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd
```

Create a random 4x4 array from an exponential distribution

```python
exp_array = np.random.exponential(scale=1.0, size=(4, 4))
print(exp_array)
```

Create a random 100000x1 array from the same exponential distribution and visualize the
distribution histogram along with uniform and normal distributions

```python
# Arrays
exp_data = np.random.exponential(scale=1.0, size=(100000,))
uni_data = np.random.rand(100000)
norm_data = np.random.randn(100000)
```

```python
# Histogram Plot
plt.hist(exp_data, bins=100, density=True, histtype='step', color='red', label='Exponential')
plt.hist(uni_data, bins=100, density=True, histtype='step', color='blue', label='Uniform')
plt.hist(norm_data, bins=100, density=True, histtype='step', color='green', label='Normal')
```

```python
plt.axis([-2.5, 10, 0, 1.5])
plt.legend(loc='upper right')
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5]

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("Z = X² + Y²")
plt.show()
```
